```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split

base_dir = '/content/drive/Shareddrives/datasci/ton_iot/result/'
```

```python
df = pd.read_csv('/content/drive/Shareddrives/datasci/ton_iot/Train_Test_Network.csv',
                      sep=',',
                      header=0,
                      dtype={'label': int},
                      usecols=['label'],
                      na_values=['-'])
```

```python
_,Y_test = train_test_split(df['label'].values,
                            test_size=0.3,
                            random_state=42)
with open(base_dir+'LABEL.csv','w') as outfile:
  aux = ''
  for y in Y_test:
    aux+=str(y)+','
  outfile.write(aux[:-1]+'\n')
```

```python
with open(base_dir+'ALL.csv','w') as outfile:
  for filename in os.listdir(base_dir):
    if filename.startswith('ALL') or not filename.endswith('.csv'):
      continue
    with open(base_dir+filename,'r') as infile:
      model_name = filename.replace('.csv','')
      outfile.write(model_name+','+next(infile))
```
